```python
def generate_tokenizer(corpus):
    test = corpus
    t0 = DefaultTagger("n")
    t1 = UnigramTagger(test, backoff=t0)
    t2 = BigramTagger(test, backoff=t1)
    return t2
```

```python
def process2(phrases):
    taggedPhrases = []
    for phrase in phrases:
        n_phrase = t.tag(phrase.split(' '))
        taggedPhrases.append(n_phrase)
    process3(taggedPhrases)
```
